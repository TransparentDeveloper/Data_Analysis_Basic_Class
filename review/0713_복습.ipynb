## groupby 메서드

```python
df.head()
```

```python
df.groupby('기업')['현재_주가'].mean().to_frame()
```

```python
df.groupby(['연도','기업'])['현재_주가'].max().to_frame()
```

```python
# 조건변수를 인덱스 사용안함
# True로 지정시, 위와 같은 결과
df.groupby(['연도','기업'],as_index=False)['현재_주가'].max()
```

## agg 메서드
- 여러가지 함수 요소 리스트를 입력하면 데이터 값이 함수 요소에 반영되어 대상 변수 값이 변함.
- 즉, 내가 원하는 변수의 통계치를 볼 수 있음

```python
import numpy as np
```

```python
df.groupby('기업',as_index=False)[['현재_주가','미래_주가']].agg([max,min,np.mean])
```

```python
df.groupby('기업',as_index=False)[['현재_주가','미래_주가']].agg({'현재_주가':[max,min], '미래_주가':[np.mean]})
```
