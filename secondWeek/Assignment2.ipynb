## 다양한 그래프 그려보기

```python
#선 그래프
plt.plot([1,2,3],[4,5,6],label='Line1')
plt.plot([1,2,3],[3,-1,7],label='Line2')
plt.legend(loc='left')
```

```python
#선 그래프, 색상 설정, 선 스타일, 두께 설정
plt.plot([1,2,3],[4,5,6],label='Line1',color='black',linestyle='dashed',linewidth=4)
plt.plot([1,2,3],[3,-1,7],label='Line2',color='red',linestyle='dotted',linewidth=3)
plt.legend(loc='left')
```

```python
# 마커 설정하는 법
plt.plot([1,2,3],[4,5,6],label='Line1',color='black',marker='s',linewidth=4,markersize=8)
plt.plot([1,2,3],[3,-1,7],label='Line2',color='red',marker='o',linewidth=3,markersize=8)
plt.legend(loc='left')
```

```python
# 마커를 이용한 산점도
plt.scatter([1,2,3],[4,1,2],marker='*',s=80,color='red',label='scatter1')
plt.scatter([1,2,3],[7,8,9],marker='o',s=80,color='blue',label='scatter2')
```

```python
x = np.random.normal(10,5,100)
plt.boxplot(x)
```

- 맨 아래 동그라미는 이상치
- 양끝 - 수염 윗쪽 (Q3+1.5*(Q3-Q1)) IQR / 아래쪽 (Q1 - 1.5*(Q3-Q1))
- 빨간색은 중위수
- 네모의 가로 위가 3Q 아래가 1Q

## 히트맵
- 2차원 배역의 각 요소를 시각화 하는 법

```python
import pandas as pd
data = pd.DataFrame({"Math":[100,90,80],'English':[70,60,50]},
                    index={'A','B','C'})
```

```python
sns.heatmap(data,annot=True,fmt='d') #annot 박스에 대응되는 값이 데이터 표시,fmt: 데이터를 어떻게 표시할지 체크 d는 정수만, 3f
```

```python
sns.heatmap(data,annot=True,fmt='d',linewidth=1,linecolor='black') #annot 박스에 대응되는 값이 데이터 표시,fmt: 데이터를 어떻게 표시할지 체크 d는 정수만, 3f
```

```python
sns.heatmap(data,annot=True,fmt='.3f',cmap='Greys') #cmap 전체 컬러의 배합을 바꾸는 것
```
