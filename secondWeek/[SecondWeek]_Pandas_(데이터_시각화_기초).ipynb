## 다양한 그래프 그려보기

```python
import seaborn as sns

from matplotlib import pyplot as plt
from matplotlib import rcParams

sns.set()
%matplotlib inline

rcParams['font.family'] = 'Malgun Gothic'
rcParams['axes.unicode_minus'] =False
```

```python
#선그래프 
plt.plot([1,2,3],[4,1,2], label='Line1')
plt.plot([1,2,3],[5,6,7], label='Lne2')
plt.legend(loc='left')
```

```python
#선그래프, 색상 설정
plt.plot([1,2,3],[4,1,2], label='Line1',color='black')
plt.plot([1,2,3],[5,6,7], label='Lne2', color='red')
plt.legend(loc='left')
```

## 과제1. 선스타일, 라인스타일, 구글링해서 새로운 것으로 그래프 그리기! 

```python
#선그래프, 색상 설정 선 스타일, 두께 설정
## 선스타일, 라인스타일, 구글링해서 새로운 것으로 그래프 그리기! 
plt.plot([1,2,3],[4,1,2], label='Line1',color='black', linestyle='dashed', linewidth=4)
plt.plot([1,2,3],[5,6,7], label='Lne2', color='red', linestyle='dotted', linewidth=3)
plt.legend(loc='left')
```

## 과제 추가 _ 마커도 다양해요 다양한 마커로 그래프 그려보기! ( 구글링 )

```python
#마커설정하는 법
plt.plot([1,2,3],[4,1,2], label='Line1',color='black', marker="s", markersize= 8)
plt.plot([1,2,3],[5,6,7], label='Lne2', color='red', marker='o', markersize= 8 )
plt.legend(loc='left')
```

```python
#마커를 이용한 산점도

plt.scatter([1,2,3],[4,1,2],marker='*', s=80, color='red',label='scatter1')
plt.scatter([1,2,3],[7,8,9],marker='o', s=80, color='blue',label='scatter2')
```

## 히트맵 
- 2차원 배열의 각 요소를 시각화 하는 법

```python
import pandas as pd
data = pd.DataFrame({'안녕':[100,80,90],'eng':[70,60,50]},
                   index={'D','B','C'})#인덱스가 반대로 들어간다.
```

```python
data
```

```python
sns.heatmap(data, annot=True, fmt='.3f') #annot 박스에 대응되는 값이 데이터 표시, fmt 데이터를 어떻게 표시할지 체크 d정수만, .3f
```

```python
sns.heatmap(data, annot=True, fmt='.3f',linewidth = 1, linecolor='black') #annot 박스에 대응되는 값이 데이터 표시, fmt 데이터를 어떻게 표시할지 체크 d정수만, .3f
```

```python
sns.heatmap(data, annot=True, fmt='.3f',cmap='Greys') #cmap 전체 컬러의 배합을 바꾸는 것
```
